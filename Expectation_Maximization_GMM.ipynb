```python
# Implementation of GMM from http://ciml.info/dl/v0_99/ciml-v0_99-ch16.pdf
def GMM(X,K):
  mean=[]
  theta=[]
  variance=[]
  z = np.ones((len(X),K))
  for k in range(K):
    mean.append(randint(int(min(X)),int(max(X))))
    variance.append(1)
    theta.append(1/K)
  N=len(X)
  previous = np.ones((len(X),K))
  mean_prime = mean
  variance_prime=variance
  while True:
    previous = copy.deepcopy(z)
    mean_prime=copy.deepcopy(mean)
    variance_prime=copy.deepcopy(variance)
    for n in range(N):
      for k in range(K):
        z[n][k] = theta[k]*pow((2 * math.pi * variance[k]),-1/2) * math.exp(-1/(2*variance[k]) * (X[n]-mean[k])**2)
      z[n] = 1/(sum(z[n,:]))*z[n]
    for k in range(K):
      theta[k] = 1/N * sum(z[:,k])
      mean[k] = np.dot(z[:,k], X) / sum(z[:,k])
      variance[k] = np.dot(z[:,k] , (X[:]-mean[k]))/sum(z[:,k])
    if np.isnan(np.sum(z)):
      break
  return previous, mean_prime, variance_prime
```

```python
actual_mean=[0.0,5.0,10.0]
each_variance=4
std_dev=each_variance**0.5
actual_variance=[each_variance, each_variance, each_variance]
calculated_mean=np.ones(3,dtype=float)
calculated_variance=np.ones(3,dtype=float)

iterations = 0
count = 0
while (iterations<100):
    try:
      
      input_data_pts_1 = np.random.normal(0,std_dev,100);
      input_data_pts_2 = np.random.normal(5,std_dev,100);
      input_data_pts_3 = np.random.normal(10,std_dev,100);

      input_data_pts = np.concatenate((input_data_pts_1,input_data_pts_2,input_data_pts_3))
      z,mean,variance = GMM(input_data_pts,3)
      mean=sorted(mean)
      for i in range (3):
        calculated_mean[i]=calculated_mean[i]+mean[i]
        calculated_variance[i]=calculated_variance[i]+variance[i]
      for i in range (150):
          index=np.where(z[i,:]==max(z[i,:]))
          
          if i < 50:
            if index[0][0]==0:
              count+=1
          elif i>=50 and i<100:
            if index[0][0]==1:
              count+=1
          elif i>=100:
            if index[0][0]==2:
              count+=1
      print("mean value of", iterations, ":", mean)
      print("variance of", iterations, ":", variance)
      print("")
      iterations=iterations+1
    
    except OverflowError:
      pass
print("Mean values: ",calculated_mean/50)
print("Variance values: ",calculated_variance/50)
for i in range (3):
  calculated_mean[i]=calculated_mean[i]/100
  calculated_variance[i]=calculated_variance[i]/100
  calculated_mean[i]=abs(calculated_mean[i]-actual_mean[i])
  calculated_variance[i]=abs(calculated_variance[i]-actual_variance[i])
accuracy_percentage=count/150

print("")
print("********************************************")
print("")
print("Actual mean values: ",actual_mean)
print("Actual Variance values: ",actual_variance)
print("")
print("Accuracy of Calculated mean values: ",calculated_mean)
print("Accuracy of Calculated Variance values: ",calculated_variance)
print("")
print("Accuracy (number of correct predictions): ",accuracy_percentage,"%")
```
